```python
# Crack open any WAV file and see its raw guts
with open("sample1.mp3", "rb") as f:
    raw = f.read()

# Print header as hex
print("=== HEADER ===")
for i in range(0, 44, 16):
    chunk = raw[i:i+16]
    hex_part  = ' '.join(f'{b:02X}' for b in chunk)
    ascii_part = ''.join(chr(b) if 32 <= b < 127 else '.' for b in chunk)
    print(f"{i:03d}: {hex_part:<48}  |{ascii_part}|")

# Decode first 10 samples manually
import struct
print("\n=== FIRST 10 SAMPLES ===")
for i in range(10):
    offset = 44 + i * 2
    b1, b2 = raw[offset], raw[offset+1]
    int_val   = struct.unpack_from('<h', raw, offset)[0]  # <h = little-endian signed int16
    float_val = int_val / 32768.0
    print(f"[{i}]  bytes: {b1:02X} {b2:02X}  →  int: {int_val:6d}  →  float: {float_val:.5f}")
```
